```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0, 4 * np.pi, 0.1)
y = np.sin(x)

plt.plot(x, y, color='green', linestyle='-', linewidth=2)
plt.xlabel('x values from 0 to 4pi')
plt.ylabel('sin(x)')
plt.title('Plot of Sine Function')
plt.grid(True)
plt.show()
```
